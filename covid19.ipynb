```python
df1 = df.drop(['Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13'], axis=1)
```

```python
df1.isnull().sum()
```

```python
df1.dtypes
```

```python
x = df1.groupby('location')['total_cases'].mean().nlargest(10)

plt.figure(figsize=(15, 8))
bars = plt.bar(x.index, x.values, color='skyblue')
plt.title('Top 10 Locations of Total Cases')
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Count Of Cases', fontsize=12)
plt.ylim(10000, 5000000)
plt.xticks(rotation=90)
plt.grid(True)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50000,f'{int(yval):,}', ha='center', va='bottom', fontsize=10)
```

```python
x = df1.groupby('location')['total_deaths'].mean().nlargest(10)
plt.figure(figsize=(15, 8))
bars = plt.bar(x.index, x.values, color='skyblue')
plt.title('Top 10 Locations of Total Deaths')
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Count Of Cases', fontsize=12)
plt.ylim(10000, 100000)
plt.xticks(rotation=90)
plt.grid(True)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10,f'{int(yval):,}', ha='center', va='bottom', fontsize=10)
```

```python
x = df1.groupby('location')['total_cases'].mean().nsmallest(10)

plt.figure(figsize=(15, 8))
bars = plt.bar(x.index, x.values, color='skyblue')
plt.title('Bottom 10 Locations of Total Cases')
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Count Of Cases', fontsize=12)
plt.ylim(0, 50)
plt.xticks(rotation=90)
plt.grid(True)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1,f'{int(yval):,}', ha='center', va='bottom', fontsize=10)
```
